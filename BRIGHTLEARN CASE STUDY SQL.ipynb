**I WANT TO SEE MY TABLE IN THE CODING TO START EXPLORING EACH COLUMN.**

```sql
SELECT*
FROM workspace.default.bright_coffee_shop_sales_2
LIMIT 10;
```

1.**CHECKING THE DATA RANGE**
- **THEY STARTED COLLECTING THE DATA ON THE 2023-01-01**
- **THEY CONCLUDED COLLECTIING DATA ON THE 2023-06-30,THE DURATION OF THE DATA IS 6 MONTHS****

```sql

SELECT MIN(TRANSACTION_DATE) AS START_DATE
FROM workspace.default.bright_coffee_shop_sales_2;


SELECT MAX(TRANSACTION_DATE) AS END_DATE
FROM workspace.default.bright_coffee_shop_sales_2;

```

2. **CHECKING NAMES OF THE DIFFERENT STORE LOCATIONS**

-**WE HAVE 3 STORES AND THEIR NAMES ARE LOWER MANHATTAN, HELL'S KITCHEN AND ASTORIA**.

```sql
SELECT DISTINCT STORE_LOCATION
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT COUNT(STORE_ID) AS NUMBER_OF_STORES
FROM workspace.default.bright_coffee_shop_sales_2;
```

3. **CHECKING FOR PRODUCTS SOLD IN OUR STORES.**
- **THIS HELPS US IDENTIFY WHAT PRODUCTS ARE SOLD IN OUR STORES**.

```sql
SELECT DISTINCT PRODUCT_CATEGORY
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT DISTINCT PRODUCT_DETAIL
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT DISTINCT PRODUCT_TYPE 
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT DISTINCT PRODUCT_CATEGORY AS CATEGORY,PRODUCT_DETAIL AS PRODUCT_NAME
FROM workspace.default.bright_coffee_shop_sales_2;

```

4. **CHECKING PRODUCT PRICES** 

-**WITH THIS INFORMATION, WE ARE ABLE TO DETERMINE THE HIGHEST SELLING AND LOWEST SELLING PRODUCTS. THIS ALSO GIVES US PEAK TIME INTERVALS FOR SALES AND TOTAL REVENUE PER PRODUCT TYPE.**

```sql
SELECT MIN(UNIT_PRICE) AS CHEAPEST_PRICE 
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT MAX(UNIT_PRICE) AS EXPENSIVE_PRICE
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT COUNT(*) AS NUMBER_OF_ROWS, COUNT(TRANSACTION_ID) AS NUMBER_OF_SALES, COUNT(PRODUCT_ID) AS NUMBER_OF_PRODUCTS, COUNT(STORE_ID) AS NUMBER_OF_STORES
FROM workspace.default.bright_coffee_shop_sales_2;

SELECT *
FROM workspace.default.bright_coffee_shop_sales_2
LIMIT 10;

SELECT TRANSACTION_ID,TRANSACTION_DATE,DAYNAME(TRANSACTION_DATE) AS DAY_NAME,MONTHNAME(TRANSACTION_DATE) AS MONTH_NAME,COUNT(DISTINCT TRANSACTION_ID) AS NUMBER_OF_SALES,SUM(TRANSACTION_QTY) AS REVENUE_PER_DAY
FROM workspace.default.bright_coffee_shop_sales_2
GROUP BY TRANSACTION_ID,TRANSACTION_DATE,DAY_NAME,MONTH_NAME;
```
